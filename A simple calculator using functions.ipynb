```python
def mainMenu():
    validChoices = ["A", "S", "D", "M", "F", "P", "R", "E"]
    valid = False
    while not valid:
        print("A)dd numbers")
        print("S)ubtract numbers")
        print("D)ivide numbers")
        print("F)loor divide numbers")
        print("M)ultiply")
        print("P)ower")
        print("R)emainder")
        print("E)xit the menu")
        choice = input("choose your option").upper()
        if choice not in validChoices:
            print("sorry that is not a valid menu choice ")
    
        else:
            valid= True
            
    return choice
    

def enterNumber():
    valid = False
    while not valid:
        try:
            x=int(input("enter the first number"))
            valid=True
        except ValueError:
            print("That was not a valid number.. please try again")
    
    return x


def add(x,y):
    
    return x+y


def subtract(x,y):
    
    return x-y


def divide(x,y):
    
    return x/y


def multiply(x,y):
    
    return x*y


def floor_division(x,y):
    
    return x//y


def power(x,y):
    
    return x**y
    
def remainder(x,y):
    
    return x%y
    
    
#main program starts here
print("welcome to my calculator")
again = "y"
while again == "y":
    
    userChoice = mainMenu()
    if userChoice == "A":
        answer = add(enterNumber(), enterNumber())
    elif userChoice == "S":
        answer = subtract(enterNumber(), enterNumber())
    elif userChoice == "D":
        answer = divide(enterNumber(), enterNumber())
    elif userChoice == "E":
        again = "N"
    else:
        print("error - we should never get here")
    
    print("the answer from your calcaulation is="+str(answer))
    
```
